#### Having a go at timeseries analysis on Glubbdubdrib's data

```python
from statsmodels.tsa.seasonal import STL
plt.rcParams['figure.figsize'] = [12,8]

### As we are going to do this multiple times, let's create a function

def decomp(df,series,title,robust=False):
    
    # Fit the seasonal decomposition
    stl = STL(df[series], robust = robust)
    result = stl.fit()
    
    # Plot the result. You do need the semicolon with statsmodels
    fig = result.plot();
    
    # Add a user-specified title to the whole chart, above the name of the
    # timeseries, which statsmodels adds automatically.
    fig.suptitle(title)
    plt.tight_layout(rect=[0, 0.03, 1, 0.99])
    
    # Extract the trend, seasonal, and residual components
    trend = result.trend
    seasonal = result.seasonal
    residual = result.resid
```

#### Let's try a Prophet forecast

```python
from prophet import Prophet
from prophet.plot import (plot_plotly, 
                          plot_components_plotly,
                          plot_forecast_component,
                          add_changepoints_to_plot)
```

```python
def prophet_training_data(y_train):
    '''
    Courtesy of Dr. Tom Monks, University of Exeter
    ---------
    Converts a standard pandas datetimeindexed dataframe
    for time series into one suitable for Prophet
    Parameters:
    ---------
    y_train: pd.DataFrame
        univariate time series data
        
    Returns:
    --------
        pd.DataFrame in Prophet format 
        columns = ['ds', 'y']
    '''
    prophet_train = pd.DataFrame(y_train.index)
    prophet_train['y'] = y_train.to_numpy()
    prophet_train.columns = ['ds', 'y']

    return prophet_train
```

```python
y_train = train_glub_overall_for_pred
y_train.index.freq = 'MS'
prophet_train = prophet_training_data(y_train)
```

```python
prophet_train.head()
```

```python
p_model = Prophet(interval_width = 0.95)
p_model.fit(prophet_train)
future = p_model.make_future_dataframe(periods = 9, freq = 'MS')
prophet_forecast = p_model.predict(future)
```

```python
# need this with date index for plotting against the prediction

test_glub_overall_for_pred = test_glub_overall.copy().set_index('REFERRAL_MONTH').drop(columns=['SET'])

test_glub_overall_for_pred.head()
```

```python
p_model.plot_components(prophet_forecast);

plt.figure(figsize=(10, 6))
p_model.plot(prophet_forecast, xlabel='Date', ylabel='Value', ax=plt.gca())
test_glub_overall_for_pred.plot(color = 'red', label = 'Test data (actuals)',ax=plt.gca()) # this works slightly differently, doesn't change the label.
plt.title('Prophet Forecast Glubbdubdrib')
plt.legend()
plt.show()
```

```python
test_glub_overall['REFERRALS_RECEIVED']
```

```python
rsme_proph = math.sqrt(mean_squared_error(test_glub_overall['REFERRALS_RECEIVED'],prophet_forecast['yhat'][-9:]))
# yhat = predicted values. Remember, Prophet creates a yhat for historic value as well, so you need to get the last
# values corresponding to the number of values in the test dataset.

print(f'Prophet Forecast RSME: {rsme_proph}')
```
